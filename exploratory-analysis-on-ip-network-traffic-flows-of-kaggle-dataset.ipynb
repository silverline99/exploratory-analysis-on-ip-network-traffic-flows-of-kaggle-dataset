```python
"""
STEP 2: FREQUENCY OF OCCURENCE OF VALUES IN "PROTOCOL NAMES" COLUMN
"""
```

```python
# Display all unique protocol names in dataset
# to get the overview about types of protocols with a possibility to discover new ideas for an analysis

ntf['ProtocolName'].unique()
```

```python
# Get the frequency of occurence of unique protocol names in dataset

ntf['ProtocolName'].value_counts()
```

```python
# Get an aggregation based on frequency of occurence of unique protocol names in dataset

ntf_protocol_name_agg = ntf['ProtocolName'].value_counts()
ntf_protocol_name_agg
```

```python
# Display top20 major categories by protocol name count of occurence

ntf_protocol_name_agg.sort_values(ascending=False).head(20)
```

```python
ntf_protocol_name_agg.head(10).plot.barh(
    figsize = (10, 10),
    title="Top 10 Most Frequent Protocol Names in dataset"
);
```

```python
"""
STEP 3: The sum of packets by protocol name
"""
```

```python
# Sum the columns "Total.Fwd.Packets" and "Total.Backward.Packets" into one column

ntf["SumOfPackets"] = ntf["Total.Fwd.Packets"] + ntf["Total.Backward.Packets"]
```

```python
# Make an aggregation based on 'ProtocolName' column

ntf_packets_aggregation = ntf.groupby("ProtocolName")
ntf_packets_aggregation = ntf_packets_aggregation.sum().sort_values(by="SumOfPackets", ascending=False)
```

```python
# Display the result of aggregation in the plot

ntf_packets_aggregation["SumOfPackets"].sort_values(ascending=False).head(10).plot.barh(
    figsize = (10, 10),
    title="Top 10 Protocol Names by the biggest count of packets in dataset",
);
```

```python
"""
STEP 4: The sum of packets length by protocol name
"""
```

```python
# Description of used columns

# Total.Length.of.Fwd.Pac
# The total quantity of bytes in the forward direction obtained from all the flow (all the packets transmitted)

# Total.Length.of.Bwd.Pa
# The total quantity of bytes in the backward direction obtained from all the flow (all the packets transmitted)
```

```python
# Sum the columns "Total.Length.of.Fwd.Packets" and "Total.Length.of.Bwd.Packets" into one column

ntf["SumOfPacketsLength"] = ntf["Total.Length.of.Fwd.Packets"] + ntf["Total.Length.of.Bwd.Packets"]

#ntf["SumOfPacketsLength"] = sum_of_packets_length
```

```python
# Make an aggregation based on 'ProtocolName' column

ntf_packets_length_aggregation = ntf.groupby("ProtocolName")
ntf_packets_length_aggregation = ntf_packets_length_aggregation.sum().sort_values(by="SumOfPacketsLength")
```

```python
# Display the result of aggregation in the plot

ntf_packets_length_aggregation["SumOfPacketsLength"].sort_values(ascending=False).head(10).plot.barh(
    figsize = (10, 10),
    title = "Top 10 Biggest Protocol names by packet length"
);
```

```python
"""
STEP 6: DATA PREPARATION FOR A TIME SERIES ANALYSIS
"""
```

```python
# Define function which will modify the string in the column Timestamp

def insert_space(string, integer):
    return string[0:10] + ' ' + string[10:]
```

```python
ntf
```

```python
# Make a string in the Timestamp column convertable to DateTime format

ntf['Timestamp'] = ntf['Timestamp'].apply(lambda x: insert_space(x, 9))
```

```python
# Convert Timestamp column to DateTime format

ntf['Timestamp'] = pd.to_datetime(ntf['Timestamp'])
```

```python
"""
STEP 7: Time series analysis
"""
```

```python
# Display the number of unique source ip addresses

ntf['Source.IP'].value_counts()
```

```python
# Make aggregation of data based on Source.IP column and count the occurence
# in order to determine which source.IP is the most frequent

ntf.groupby('Source.IP').count().sort_values('Flow.ID', ascending=False)
```

```python
# Select only rows related to the particular IP address (the most frequent by occurence was choosen)

selected_ip = ntf[ntf['Source.IP']=='10.200.7.218']
selected_ip
```

```python
# Select only relevant columns

selected_ip_ts = selected_ip[["Timestamp",
                              "SumOfPackets",
            ]]
selected_ip_ts
```

```python
# Set the column "Day" as index

selected_ip_ts = selected_ip_ts.set_index("Timestamp").sort_index()
```

```python
# Group data into a one-day period

selected_ip_ts_grouped = selected_ip_ts.groupby(pd.Grouper(freq='D')).sum()
```

```python
# Display the days when Source.IP was active

selected_ip_ts_grouped.sort_values('SumOfPackets', ascending=False).head(50)
```

```python
selected_ip_ts_grouped.plot(subplots=True, figsize=(12, 9));
```
